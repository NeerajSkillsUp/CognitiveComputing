Ques 1 : Ask the user to enter a value (e.g., M) for a mathematical function.
(a) Generate x values from -10 to 10 using np.linspace().
(b) Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)

```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))
x = np.linspace(-10,10,400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(10,6))
plt.plot(x,y1,'b-',linewidth=2,label=f'{M}·x²')
plt.plot(x,y2,'r--',linewidth=2,label=f'{M}·sin(x)')

plt.title(f'Plot of y = {M}·x² and y = {M}·sin(x)')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid()

plt.show()              # remember optional for notebooks
```

Ques 2 : Create a dataset of five subjects and scores.
(a) Convert it into a Pandas DataFrame.
(b) Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math','Physics','Chemistry','Biology','Computer Science'],
    'Score': [85,78,92,88,95]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10,6))
ax = sns.barplot(x='Subject',y='Score',data=df,palette='viridis')

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',
                (p.get_x()+p.get_width()/2.,p.get_height()),
                ha='center',va='center',
                xytext=(0,10),
                textcoords='offset points')

plt.title('Subject Scores',fontsize=16)
plt.xlabel('Subjects',fontsize=12)
plt.ylabel('Scores',fontsize=12)
plt.grid(axis='y')

plt.show()
```

Ques 3 : Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
(a) Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
(b) Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(14)
data = np.random.randn(50)

fig,axes = plt.subplots(2,2,figsize=(12,10))

# Plot 1: Line plot showing cumulative sum
axes[0,0].plot(np.cumsum(data),color='blue')
axes[0,0].set_title('Cumulative Sum of Random Data')
axes[0,0].set_xlabel('Index')
axes[0,0].set_ylabel('Cumulative Sum')
axes[0,0].grid()

# Plot 2: Scatter plot with random noise
x = np.arange(50)
noise = np.random.randn(50)*0.2
axes[0, 1].scatter(x,data+noise,color='red',alpha=0.7)
axes[0, 1].set_title('Scatter Plot with Random Noise')
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Value')
axes[0, 1].grid()

plt.show()
```

Ques 4 : Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply 'seaborn' library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
sales_data = pd.read_csv(url)

# 1. Total profit of all months - line plot
plt.figure(figsize=(10,6))
sns.lineplot(x='month_number',y='total_profit',data=sales_data,
             marker='o',color='purple',linewidth=2)
plt.title('Total Profit of All Months',fontsize=16)
plt.xlabel('Month Number',fontsize=12)
plt.ylabel('Total Profit',fontsize=12)
plt.grid()
plt.xticks(sales_data['month_number'])
plt.show()

# 2. All product sales data - multiline plot
plt.figure(figsize=(12,7))
products = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
for product in products:
    sns.lineplot(x='month_number',y=product,data=sales_data,
                 marker='o',label=product,linewidth=2)

plt.title('Monthly Sales of All Products',fontsize=16)
plt.xlabel('Month Number',fontsize=12)
plt.ylabel('Sales Units',fontsize=12)
plt.grid()
plt.legend(title='Products')
plt.xticks(sales_data['month_number'])
plt.show()

# 3. Bar chart for all features
# Exclude month_number and total_profit for this plot
features = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer','total_units','total_profit']

plt.figure(figsize=(14,8))
sales_data[features].mean().plot(kind='bar',color=sns.color_palette('husl',len(features)))
plt.title('Average Values of All Features',fontsize=16)
plt.xlabel('Features',fontsize=12)
plt.ylabel('Average Value',fontsize=12)
plt.grid(axis='y')
plt.show()
```
